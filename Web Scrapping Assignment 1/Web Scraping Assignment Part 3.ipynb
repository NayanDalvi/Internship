# Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
team_page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
team_page
```

```python
soup_team = BeautifulSoup(team_page.content)
```

```python
First_Name = []
First_Name = soup_team.find("tr",class_="rankings-block__banner")
First_Name = First_Name.text.split("\n")
First_Name_df = pd.DataFrame(First_Name).transpose()
First_Name_df= First_Name_df.drop(columns = [1,2,3,5,6,9,11,12,13,14])
First_Name_df.columns = ["Serial Number","Team","Matches","Points","Rating"]
First_Name_df
```

```python
Team_other_9 = []
for i in soup_team.find_all("tr",class_="table-body"):
    Team_other_9.append(i.text.split("\n"))
Team_other_9 = Team_other_9[0:9]
Team_other_9
```

```python
Team_other_9_df = pd.DataFrame(Team_other_9)
Team_other_9_df = Team_other_9_df.drop(columns = [1,2,3,5,6,10])
Team_other_9_df.columns = ["Serial Number","Team","Matches","Points","Rating"]
Team_other_9_df
```

```python
Team_Output = pd.concat([First_Name_df,Team_other_9_df],ignore_index=True)
Team_Output
```
