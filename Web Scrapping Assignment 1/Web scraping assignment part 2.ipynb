# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
Name = []
Matches = []
Points = []
Rating = []
```

```python
soup = BeautifulSoup(page.content)
```

```python
All = []
for i in soup.find_all("tr",class_="table-body"):
    All.append(i.text.split("\n"))
All_9 = All[0:9]
All_9
```

```python
df = pd.DataFrame(All_9, columns = ["Serial Number","1","2","3","Name","5","6","Matches","Points","Rating","10"])
df
```

```python
tr = df.drop(columns = ["1","2","3","5","6","10"])
tr
```

```python
a = []
first = []
first_list = []
first = soup.find("tr",class_="rankings-block__banner")
a = first.text.split("\n")
df = pd.DataFrame(a)
dfx = df.transpose()
dfxx = dfx.drop(columns = [1,2,3,5,6,9,11,12,13,14])
dfxx.columns = ["Serial Number","Name","Matches","Points","Rating"]
dfxx
```
